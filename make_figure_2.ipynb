```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

x = data.index

fig=plt.figure()
# figure(figsize=(20, 10))
fig.show()
ax=fig.add_subplot(111)

ax.plot(data.index,data.A,c='b',label='Subject A',fillstyle='none')
ax.plot(data.index,data.B,c='g',label='Subject B')
ax.plot(data.index,data.C,c='k',label='Subject C')

fig.set_figheight(10)
fig.set_figwidth(20)
plt.xticks(rotation = 90)
plt.legend(loc=2)
plt.show()
```

# Figure 1. Example Data 
Following the data dissemination pattern presented in our manuscript, we begin by describing the figure created in this notebook. 

The data used in this figure is simulated blood pressure data for three patients with measurements spanning two years. We use this simulated data simply to show the interaction between the data file (in our repository at ~/data/fakeBP.csv), the parsing methods (in our repository at ~/load_data.py) and the scripts used to make figures (this script). Note in code cell 1 how we import the parsing script, then in cell 2 we can call the load_from_box method therein. 

We create a longitudinal graph to show how each simulated subject's blood pressure changes over time.

First, we import our loader module. This brings the functions defined there (in our repository at ~/load_data.py) into scope so we can use them in this script.

```python
import load_data
```

Now, we load our data. Here we call the <code>load_from_file</code> function defined in <code>load_data</code> which returns the blood pressure data. We store it as <code>data</code>.

Calling <code>head</code> shows the first several lines of the dataframe, which provides an idea of the type of data present and the structure of the dataframe.

```python
data = load_data.load_from_file()
data.head()
```

Once we have our data, we can analyze it. Here, we make a simple longitudinal plot of the three fake subjects' readings. This plot shown in our manuscript's Figure 1, a diagram that displays the layout of our data analysis dissemination pattern.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,5))
sns.set(style="ticks", rc={"lines.linewidth": 5})
fig = sns.lineplot(data=[data['Subject 1'], data['Subject 2'], data['Subject 3']],
                   dashes = False, markers = [".",".","."])
fig.set(ylabel="Blood Pressure (mm Hg)")#xlabel="Date",
#plt.title("Fake Blood Pressures Over Time", fontsize = 20)
plt.xticks(rotation=45,horizontalalignment='right')
plt.savefig('data/Figure1.png',dpi=300)
plt.show()
```

This code saves the figure as ~/data/Figure1.png in the same folder as the repository. The figure is then used in the manuscript to illustrate scientific conclusions, and the availability of the code leading to that figure ensures the analysis is easily reproducible and transparent as it is disseminated alongside the publication.

Following the data dissemination pattern presented in our manuscript, we begin by describing the figure created in this notebook. 

The data used here is stored on Box to demonstrate how data may be accessed if it cannot be uploaded to the repository, such as with files that are too large to be stored on GitHub or that need to be password protected. The file is retrieved from box using the download url (in our repository at ~/data/url.txt) to stream the data file (which is stored locally following data at ~/data/datafile.csv).

The data used in this figure is simulated blood pressure data for three patients with measurements spanning two years. We are using this simulated data simply to show the interaction between the data file, the parsing methods (in our repository at ~/load_data.py) and the scripts used to make figures (this script). Note in code cell 1 how we import the parsing script, then in cell 2 we can call the load_from_file method therein.

Here we create two graphs: first, a longitudingal graph of Subject 1; and second, a boxplot to show how the subjects compare overall.

First, we import our loader module. This brings the functions defined there (in our repository at ~/load_data.py) into scope so we can use them in this script.

```python
import load_data
```

Next, load the data. Unlike in make_figure1, where the data is loaded from a local file stored with the repository, here we will load it from box using <code>load_from_box</code>. This function downloads a remote copy of the data file and then returns the data. This may be useful in the case of files too large to be stored on GitHub. The function gets the url from a local text file (in our repository at ~/data/url.txt) which contains only a download url: https://byu.box.com/shared/static/xzki9ynv82frwdrg2gcrcdnqcivtcjvq.csv

Calling <code>head</code> shows the first several lines of the dataframe, which provides an idea of the type of data present and the structure of the dataframe.

```python
data = load_data.load_from_box()
data.head()
```

Then, proceed with your analysis. First, we create a longitudingal graph from the Subject 1's data using <code>matplotlib.pyplot</code> tools.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 5))
plt.title("Fake Blood Pressure Subject 1")
x = data.index
y = data['Subject 1']
plt.plot(x, y)
plt.xticks(rotation=45,horizontalalignment='right')
plt.show()
```

Next, we compare the subjects with boxplots. This gives an idea of the variation in each set of measurements. While exploring your data, you may analyze it in a variety of ways. This exploration will be easy to follow and repeat using the pattern.

```python
import pandas as pd
import seaborn as sns

fig = sns.boxplot(x="variable", y="value", data=pd.melt(data))
fig.set(xlabel="",ylabel="Blood Pressure (mm Hg)", title="Fake Blood Pressures")

plt.show()
```
